#code for get total amt

```python

def getTotalAmount(input_str):
  import re

  res = input_str.split('\sor\s')

  for z in res:
      string = z.split(".)")
      i = string[0] + '.'
      # print("i is : " + i)

      if re.findall("\d*\s*\d+\/\d+", i):
        parts = re.findall("\d*\s*\d+\/\d+", i)
        for part in parts:
          # print("part : " + part)
          sep_parts = part.strip().split(" ")
          # print("seperated parts length: ", len(sep_parts))

          sum = 0;

          if len(sep_parts) == 1:
            fraction =  sep_parts[0].split("/")
            sum = int(fraction[0].strip()) / int(fraction[1].strip())
            # print ("sum : " , sum)

          if len(sep_parts) == 2:
            fraction =  sep_parts[1].split("/")
            f = int(fraction[0].strip()) / int(fraction[1].strip())
            w = int( sep_parts[0].strip())
            sum = w + f
            # print ("sum : " , sum)

          sum_string = str(sum)
          i_frac = i
          i = i.replace(part, str(sum))
          input_str = input_str.replace(i_frac, i)
          # print(i)

      digits = re.findall("\d+\.\d+|\d+" , i)
      # print (digits)
      total = 0
      if len(digits) == 2:
        total = float(digits[0]) * float(digits[1]);

      #print(total)
      final = re.sub('\d+\.\d\s\(\d+\.\d+|\d+\.\d\s\(\d+|\d+\s\(\d+\.\d+|\d+\s\(\d+', str(total), i)
      final = re.sub('\d+\.*\d*\s\(\d+\.*\d*', str(total), i)
      #print('Final : ' + final)
      #print("i is currently : " + i)

      input_str_orig = input_str
      input_str = input_str.replace(i , final)
      input_str = input_str.translate({ord(')'): None})
      # print('replaced : ' + input_str)
      #print('new test string is : ' + input_str)
  # print('original test string is : ' + input_str_orig)
  # print('new test string is : ' + input_str)
  return input_str
```
