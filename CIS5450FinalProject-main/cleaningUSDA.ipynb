```python
def removeSpecialChar(each):
  import re
  each = str(each)
  each = each.replace('"', '')
  each=each.replace("[","")
  each=each.replace("]","")
  each=each.replace("(","")
  each=each.replace(")","")
  each=each.replace("-"," ")
  each=each.replace(".","")
  final_each=each.strip()
  return final_each
```

#food_portion csv

```python
food_portions_path = "/content/drive/MyDrive/recipe project/FoodData_Central_csv_2023-04-20 (1)/food_portion.csv"

df_food_portions = pd.read_csv(food_portions_path)
df_food_portions
```

```python
df_food_portions.drop(["id", "seq_num", "portion_description", "data_points",	"footnote",	"min_year_acquired" ], axis=1)
```

```python
import numpy as np
df_food_portions['measure_unit_id'] = df_food_portions['measure_unit_id'].replace(9999,np.NaN)
```

```python
df_food_portions['portion_description'] = df_food_portions['portion_description'].replace('Quantity not specified',np.NaN)
```

```python
display(df_food_portions)
```

```python
df_food_portions_final = df_food_portions.drop(["id", "seq_num", "data_points",	"footnote",	"min_year_acquired" ], axis=1)
```

```python
display(df_food_portions_final)
```

```python
df_food_portions_final['modifier'] = df_food_portions_final['modifier'].where(df_food_portions_final['modifier'].str.match(r'\D+$', na=False))
```

```python
display(df_food_portions_final)
```

```python
df_food_portions_final_1 = df_food_portions_final.dropna(subset=['measure_unit_id', 'portion_description', 'modifier'], how='all')
```

```python
display(df_food_portions_final_1)
```

```python
df_food_portions_final_2 = df_food_portions_final_1
df_food_portions_final_2['amount'] = df_food_portions_final_1['amount'].fillna("1.0")
```

```python
display(df_food_portions_final_2)
```

```python
df_food_portions_final_2 = df_food_portions_final_2.replace({'portion_description' : {'tablespoon' : 'tbsp', 'teaspoon' : 'tsp', 'cup' : 'c', 'ounce' : 'oz', 'pint' : 'pt', 'quart' : 'qt', 'package' : 'pkg', 'pound' : 'lb'}}, regex=True)
df_food_portions_final_2 = df_food_portions_final_2.replace({'modifier' : {'tablespoon' : 'tbsp', 'teaspoon' : 'tsp', 'cup' : 'c', 'ounce' : 'oz', 'pint' : 'pt', 'quart' : 'qt', 'package' : 'pkg', 'pound' : 'lb'}}, regex=True)
```
